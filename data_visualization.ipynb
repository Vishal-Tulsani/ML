```
x1=[2,6,8,10,12]
y1=[3,6,9,12,15]
x2=[12,16,18,16,11]
y2=[13,16,19,1,5]
```

```
plt.xlabel("Distance")  # x axis label
plt.ylabel("Time")   # y axis label
plt.grid(c="violet")  # to put on graph lines
plt.plot(x1,y1,label="cars")
plt.plot(x2,y2,label="bikes")
plt.legend()  # to show the labels of plot
```

BAR PLOTS

```
plt.xlabel("asdf")
plt.ylabel("qwer")
plt.grid(c="orange")
plt.bar(x1,y1,label="Samsung")
plt.bar(x2,y2,label="microsoft")
plt.plot(x1,y1,label="Amazon",c="seagreen")
plt.legend()
```

Scatter / Dots plot

```
plt.scatter(x1,y1,marker='*',s=100)
plt.scatter(x2,y2,marker='^',s=200)
plt.grid(c='y')
```
